```python
df = pd.read_csv("/content/drive/Shareddrives/솔룩스 데이텀/naver_book_책정보.csv")
df
```

```python
df = df[df['1'] != 'X']
df = df.reset_index()
df
```

```python
df = df.drop(['index'], axis=1)
df
```

```python
# 경고문을 무시합니다.
import warnings
warnings.filterwarnings(action='ignore')

from konlpy.tag import Okt

okt = Okt()
```

```python
stem_pos_results=[]

#3974
for i in range(0,len(df['0']),10):
  list=[]
  list = okt.pos(df['1'][i], stem =True)
  stem_pos_results.append([df['0'][i],list])

print(stem_pos_results)
```

```python
print(len(stem_pos_results))
```

```python
stem_df = pd.DataFrame(stem_pos_results)
stem_df
```

```python
print(stem_pos_results[0][1][0])
```

```python
result=[]
for i in range(len(stem_pos_results)):
  list=[]
  for j in range(len(stem_pos_results[i][1])):
    a,b = stem_pos_results[i][1][j]
    if b == 'Noun' or b == 'Adjective' or b == 'Verb':
      list.append(a)
      #result.append(list)
    else:
      continue
  result.append([stem_pos_results[i][0], list])
```

```python
print(len(result))
```

```python
result_df = pd.DataFrame(result)
result_df
```
